Programa para exibir o fatorial de um número escolhido pelo usuário. Solução por meio de recursividade indireta.


```python
def fatorial(num):
  if num == 1:
    print(num)
    return num
  else:
    print(f'{num} x ', end = "")
    return num * fatorial(num-1)


num = int(input("Digite um número seu macaco nojento:"))
print(f'{num}! =' end="")
fat = fatorial(num)
print(fat)
```

fazer um programa parea exibir a sequência de fibonacci com o limite de elementos sendo passado pelo usuário

```python
def sequenciarFibonacci(n):
  if n==1 or n==2:
    return 1
  else:
    return sequenciarFibonacci(n - 1) + sequenciarFibonacci(n - 2)


numElements = int(input("Digite quantos elementos você quer ver na sequência de fibonacci filha da puta:"))
for i in range(numElements):
  print(sequenciarFibonacci(i+1), end = "")
```
