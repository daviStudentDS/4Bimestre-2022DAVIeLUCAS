```python
nome = input('Nome do aluno:')
n1 = float(input("N1:"))
n2 = float(input("N2:"))
n3 = float(input("N3:"))
media = (n1 + n2 + n3)/3

print(f"A média do aluno {nome} e: {media:.58f}")

if media > 6:
  print('APROVADO PRR')
else:
  print("REPROVADO BURRO")
```
